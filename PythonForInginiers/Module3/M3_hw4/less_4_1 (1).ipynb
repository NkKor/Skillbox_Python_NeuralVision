```python
import keras
from keras.layers import *
import numpy as np
import matplotlib.pyplot as plt
import keras.utils

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
input_shape = (28, 28, 1)

num_classes = 10  # Number of classes (digits 0-9)
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

model_2 = keras.Sequential([
    keras.Input(shape = input_shape),
    Flatten(),
    Dense(55,activation='relu'),
    Dense(25,activation='relu'),
    Dense(15,activation='relu'),
    Dense(10,activation='softmax')
])
model_2.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
model_2.summary()
```

```python
model_2.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
history = model_2.fit(x_train, y_train, epochs=10, batch_size=32, validation_data= (x_test, y_test))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Точность модели')
plt.xlabel('Точность')
plt.ylabel('Кол. эпох')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Потери модели')
plt.xlabel('Потери')
plt.ylabel('Кол. эпох')
plt.show()
```
