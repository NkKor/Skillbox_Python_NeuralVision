```python
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y = np.zeros((x.shape[0],2))
y[:,0] = np.sin(x)
y[:,1] = np.cos(x)
plt.plot(x, y,'.-') # без '.-' график рисуется тонкой линией
plt.title("Граффики функций sin и cos")
plt.legend(['sin (x)','cos (x)'])
plt.xlabel('ось x')
plt.ylabel('ось y')
```

```python
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
y = np.sin(x[::3]+0.3)
plt.plot(x[::3], y,'.-') # без '.-' график рисуется тонкой линией
plt.title("Граффики функций sin")
plt.legend(['sin (x)','cos (x)'])
plt.xlabel('ось x')
plt.ylabel('ось y')
plt.grid(visible=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_gen = lambda c,s: np.random.normal(loc=c, scale=s, size=100)
data_gen(0,1)
 # array([ 2.1065401 , 0.92460846, -0.09477523, -0.2424704 ,
 # 0.8866251 , -0.30897488, 1.5395672 , 0.1655203 , ...)

centers = [-2, 0, 2]
scales = [1, 2, 3]
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle("Гистограммы выборки из нормального распределения")
for i, center in enumerate(centers):
    for j, scale in enumerate(scales):
        axes[i, j].hist(data_gen(center, scale), bins=20) 
        axes[i, j].set_xlim(-10, 10) 
        axes[i, j].set_title(f"Центр: {center}. Дисперсия {scale}")
```

```python
import numpy as np
data = [2, 4, 2, 4, 1, 5, 1, 2, 5, 2]
print("Среднее значение:", np.mean(data))
print("Медиана:", np.median(data))
# Среднее значение: 2.8
# Медиана: 2.0
import scipy as sp
print("Мода:", sp.stats.mode(data))
# Мода: ModeResult(mode=2, count=4)
```

```python
# Меры разброса
print("Минимальное значение:", np.min(data))
print("Максимальное значение:", np.max(data))
# Минимальное значение: 1
# Максимальное значение: 5
print("Размах выборки:", np.max(data)-np.min(data))
# Размах выборки: 4
print("Межквартильный размах выборки:", sp.stats.iqr(data))
sorted(data)
print("Дисперсия:", np.var(data, ddof=1))
print("Среднеквадратичное отклонение:", np.std(data, ddof=1))
```

```python
import plotly.graph_objects as go

trace1 = go.Scatter(
  x=[1, 2, 3, 4],
  y=[10, 11, 12, 13],
  mode="markers",
  name="Trace 1",
  marker=dict(color="rgba(255, 0, 0, 0.8)", size=5),
 )

trace2 = go.Scatter(
  x=[1, 2, 3, 4],
  y=[12, 13, 14, 15],
  mode="lines",
  name="Trace 2",
  line=dict(color="rgba(0, 255, 0, 0.8)", width=2),
 )

data = [trace1, trace2]
layout = go.Layout(
  title="Example Graph",
  xaxis=dict(title="X Axis"),
  yaxis=dict(title="Y Axis"),
 )

fig = go.Figure(data=data, layout=layout)
fig.show() 
```
