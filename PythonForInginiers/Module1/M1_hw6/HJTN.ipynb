```python
import socket
import sys
from loguru import logger


SERVER_ADDRESS = ('localhost', 5001)
BUFFER_SIZE = 16

def init_socket():
    # Create a TCP/IP socket
    # socket.AF_INET - указывает на address family ip v4
    # socket.SOCK_STREAM - cnhb
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    # Привязываем адрес и порт
    logger.info('starting up on {:s} port {:d}' .format(*SERVER_ADDRESS))
    sock.bind(SERVER_ADDRESS)
    return sock

def serve_connections(sock):
    # Активируем
    # 1 - означает, сколько подключений может ожидать в очереди до того, как их приняли.
    sock.listen(1)

    while True:
        # Ожидаем подключение
        logger.info('waiting for a connection')
        connection, client_address = sock.accept()

        try:
            logger.info(f'connection from {client_address}')
            # Receive the data in small chunks and retransmit it
            while True:
                data = connection.recv(BUFFER_SIZE)
                logger.info(f'received "{data}"')
                if data:
                    logger.info('sending data back to the client')
                    connection.sendall(data)
                else:
                    print('no more data from', client_address)
                    break

        finally:
            # Clean up the connection
            connection.close()

def main():
    sock = init_socket()
    serve_connections(sock)
    sock.close()

if __name__ == '__main__':
    main()

```

```python
import socket
from server_socket import SERVER_ADDRESS, BUFFER_SIZE

HOST, PORT = SERVER_ADDRESS
print('Адрес сервера "{}" порт {}'.format(HOST, PORT))

# Адрес сервера "localhost" порт 5001

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

print('Подключаемся к серверу {}:{}'.format(*SERVER_ADDRESS))
sock.connect(SERVER_ADDRESS)


try:
    # Отправляем данные
    message = 'Hello world! It is the first message'
    print(f'sending "{message}"')
    sock.sendall(message.encode('ascii'))

# Нам необходимо понимать, как долго нам читать сообщения
# Должен быть фиксированный размер сообщений
# Или мы в заголовке должны передать размер тела сообщения
# Или в конце должен быть суффикс, который указывает, что мы окончили приём
    amount_received = 0
    amount_expected = len(message)
    full_message = b''

    while amount_received < amount_expected:
        data = sock.recv(BUFFER_SIZE)
        amount_received += len(data)
        print(f'received {data}')
        full_message += data

    print(f'Итоговое сообщение "{full_message.decode("ascii")}"')

finally:
    print('closing socket')
    sock.close()
```
