## Задача 1


$F(x) = \left\{\begin{matrix}
0, \text{если } x < 0 \\
\frac{x^2}{2}, \text{если } 0 \leq x \leq \sqrt{2}\\
1, \text{если }  x > \sqrt{2}
\end{matrix}\right.$

1. Постройте график данной функции распределения.
2. Найдите вероятность того, что Х примет значение, принадлежащее интервалу $X \in (0.2;1)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Область определения
x = np.linspace(-1, 2, 500)
F = np.piecewise(x,
    [x < 0, (x >= 0) & (x <= np.sqrt(2)), x > np.sqrt(2)],
    [0, lambda x: x**2 / 2, 1]
)

# График
plt.figure(figsize=(8, 5))
plt.plot(x, F, lw=2, color='royalblue')
plt.title('Функция распределения $F(x)$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$F(x)$', fontsize=12)
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', lw=1)
plt.axvline(np.sqrt(2), color='gray', linestyle='--', lw=1, label='$x=\\sqrt{2}$')
plt.legend()
plt.show()
```

## Задача 2

Постройте данные функции и определите, являются ли они плотностью распределения некоторой случайной величины:
1. $f(x) = \dfrac{x}{\pi(1+x^2)},\ x \in (-\inf,+\inf)$
2. $f(x) = \left\{\begin{matrix}
\dfrac{1}{2},\ x \in (-1;1] \\
0,\ x \notin (-1;1] \end{matrix}\right.$
3. $f(x) = \left\{\begin{matrix}
0, \text{ если } x < 0\  \&\  x > 2 \\
ax^2, \text{ если } 0 \leq x \leq 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Настройка стиля графиков для лучшего восприятия
plt.style.use('seaborn-v0_8-whitegrid')

# Создаем фигуру с тремя расположенными друг под другом графиками
fig, axs = plt.subplots(3, 1, figsize=(8, 13))
fig.suptitle('Анализ функций на соответствие свойству плотности распределения', fontsize=16)

# --- 1. Функция f(x) = x / (π * (1 + x^2)) -------------------------------------------------------------------------

x1 = np.linspace(-5, 5, 400)
y1 = x1 / (np.pi * (1 + x1**2))

axs[0].plot(x1, y1, color='red', label=r'$f(x) = \frac{x}{\pi(1+x^2)}$')
# Добавляем горизонтальную линию на y=0 для наглядности
axs[0].axhline(0, color='black', linewidth=0.8, linestyle='--')
axs[0].fill_between(x1, y1, where=(y1 < 0), color='red', alpha=0.3)
axs[0].set_title(r"Функция 1: Не является ФПР", fontsize=12)
axs[0].set_xlabel("x")
axs[0].set_ylabel("f(x)")
axs[0].legend()
axs[0].text(-4.8, -0.1, 'Вывод: f(x) < 0\nдля x < 0',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))


# --- 2. Функция f(x) = 1/2 на [-1, 1] -------------------------------------------------------------------------------

x2 = np.linspace(-3, 3, 400)
# Задаем функцию через условия
y2 = np.where((x2 >= -1) & (x2 <= 1), 0.5, 0)

axs[1].plot(x2, y2, color='green', label=r'$f(x) = 1/2$ для $x \in [-1; 1]$')
# Заливаем площадь под кривой
axs[1].fill_between(x2, y2, color='green', alpha=0.3)
axs[1].set_title(r"Функция 2: Является ФПР (равномерное распределение)", fontsize=12)
axs[1].set_xlabel("x")
axs[1].set_ylabel("f(x)")
axs[1].legend()
# Добавляем текст с расчетом площади
area_text = r"$\int_{-1}^{1} \frac{1}{2} dx = 1$"
axs[1].text(-2.9, 0.4, f'Вывод:\nf(x) ≥ 0\nПлощадь = 1\n({area_text})',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))


# --- 3. Функция f(x) = ax^2 на [0, 2] ---------------------------------------------------------------------------------------------------

# Мы вычислили, что a = 3/8
a = 3/8
x3 = np.linspace(-1, 3, 400)
# Задаем функцию через условия
y3 = np.where((x3 >= 0) & (x3 <= 2), a * x3**2, 0)

axs[2].plot(x3, y3, color='blue', label=r'$f(x) = \frac{3}{8}x^2$ для $x \in [0; 2]$')
# Заливаем площадь под кривой
axs[2].fill_between(x3, y3, where=((x3 >= 0) & (x3 <= 2)), color='blue', alpha=0.3)
axs[2].set_title(r"Функция 3: Является ФПР при $a=3/8$", fontsize=12)
axs[2].set_xlabel("x")
axs[2].set_ylabel("f(x)")
axs[2].legend(loc='upper left')
# Добавляем текст с расчетом площади
area_text3 = r"$\int_{0}^{2} \frac{3}{8}x^2 dx = 1$"
axs[2].text(-0.9, 1.0, f'Вывод:\nПри a=3/8, f(x)≥0\nПлощадь = 1\n({area_text3})',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))

# Оптимизируем расположение элементов
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Задача 1

Известно, что случайная величина $X$ задана нормальным распределением $N(3,2)$. Найдите:

1. $p\{-3 <X<5\}$
2. $p\{X \leq 4\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- 1. Параметры распределения ---
mu = 3      # Математическое ожидание
sigma = 2   # Среднеквадратическое отклонение

# Создаем "замороженный" объект распределения для удобства
dist = norm(loc=mu, scale=sigma)

# --- 2. Вычисления ---

# P(-3 < X < 5) = P(X < 5) - P(X < -3)
p_between = dist.cdf(5) - dist.cdf(-3)

# P(X <= 4)
p_le_4 = dist.cdf(4)

# Вывод результатов
print(f"Параметры распределения: μ = {mu}, σ = {sigma}")
print("-" * 30)
print(f"1. P(-3 < X < 5) = {p_between:.4f}")
print(f"2. P(X <= 4) = {p_le_4:.4f}")
print("-" * 30)


# --- 3. Визуализация ---

# Создаем фигуру с двумя графиками
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle(f'Нормальное распределение N(μ={mu}, σ={sigma})', fontsize=16)

# Диапазон значений для оси X (обычно μ ± 4σ достаточно)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_y = dist.pdf(x)

# --- График для P(-3 < X < 5) ---
axs[0].plot(x, pdf_y, 'b-', label='Плотность вероятности')
axs[0].set_title(r'$P(-3 < X < 5)$', fontsize=14)

# Заливка области между -3 и 5
x_fill_between = np.linspace(-3, 5, 200)
y_fill_between = dist.pdf(x_fill_between)
axs[0].fill_between(x_fill_between, y_fill_between, color='skyblue', alpha=0.7)

axs[0].axvline(x=-3, color='r', linestyle='--')
axs[0].axvline(x=5, color='r', linestyle='--')
axs[0].text(0, 0.1, f'Площадь = {p_between:.4f}', fontsize=12, horizontalalignment='center')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Плотность f(x)')
axs[0].grid(True)

# --- График для P(X <= 4) ---
axs[1].plot(x, pdf_y, 'b-', label='Плотность вероятности')
axs[1].set_title(r'$P(X \leq 4)$', fontsize=14)

# Заливка области левее 4
x_fill_le = np.linspace(mu - 4*sigma, 4, 200)
y_fill_le = dist.pdf(x_fill_le)
axs[1].fill_between(x_fill_le, y_fill_le, color='lightgreen', alpha=0.7)

axs[1].axvline(x=4, color='r', linestyle='--')
axs[1].text(0, 0.1, f'Площадь = {p_le_4:.4f}', fontsize=12, horizontalalignment='center')
axs[1].set_xlabel('X')
axs[1].grid(True)


plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

## Задача 2

Случайная величина $Т$ — время работы радиолампы. Она имеет экспоненциальное распределение.

Найдите вероятность того, что лампа проработает не менее $800$ часов, если среднее время работы лампы — $400$. Постройте график плотности распределения и функции распределения.

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
mu = 400                      # среднее время
lmbd = 1 / mu                 # параметр λ
t_value = 800                 # интересующий порог

# Вероятность
P_more_800 = np.exp(-lmbd * t_value)
print(f"Вероятность того, что лампа проработает не менее 800 часов: {P_more_800:.4f}")

# Диапазон времени для графиков
t = np.linspace(0, 2000, 500)

# Плотность распределения
pdf = lmbd * np.exp(-lmbd * t)

# Функция распределения
cdf = 1 - np.exp(-lmbd * t)

# Визуализация
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

# График плотности
axs[0].plot(t, pdf, label='Плотность f(t)', color='blue')
axs[0].fill_between(t, pdf, 0, where=(t >= t_value), color='red', alpha=0.3, label='t ≥ 800')
axs[0].set_title("Плотность распределения")
axs[0].set_xlabel("Время работы (часы)")
axs[0].set_ylabel("Плотность вероятности")
axs[0].legend()
axs[0].grid(True)

# График функции распределения
axs[1].plot(t, cdf, label='F(t)', color='green')
axs[1].axvline(t_value, color='red', linestyle='--', label='t = 800')
axs[1].set_title("Функция распределения")
axs[1].set_xlabel("Время работы (часы)")
axs[1].set_ylabel("Вероятность")
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

## Задача 1

В соревнованиях по стрельбе из лука два участника. Матч длится два раунда, засчитывается последний. Каждый участник сделал по 20 выстрелов. Очки считаются по сумме выстрелов.

На основе полученных данных у первого стрелка вышло распределение Рэлея (отклонение от попадания в яблочко $10$) $Ray(0,3)$, у другого — $Ray(5,1)$. Оба лучника в отличной форме и за один раунд лучше стрелять не научатся.

У кого шанс выиграть выше и что нужно для этого сделать данному лучнику?

Необходимо:
1. Построить результаты распределения после первого раунда на одном графике.
2. Описать, кто выиграет и что для этого нужно сделать данному лучнику.
3. Проверить гипотезу численно, сгенерировав полученный результат. Для упрощения считаем, что можно получать дробные числа.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

# Параметры стрелков
loc1, scale1 = 0, 3
loc2, scale2 = 5, 1
shots_per_round = 20
rounds = 100000  # количество симуляций

# Симуляция отклонений (меньше — лучше)
deviation1 = rayleigh.rvs(loc=loc1, scale=scale1, size=(rounds, shots_per_round))
deviation2 = rayleigh.rvs(loc=loc2, scale=scale2, size=(rounds, shots_per_round))

# Переводим отклонения в очки (максимум 10)
score1 = np.clip(10 - deviation1, 0, 10).sum(axis=1)
score2 = np.clip(10 - deviation2, 0, 10).sum(axis=1)

# Вероятность победы первого
p1_win = np.mean(score1 > score2)
p2_win = np.mean(score2 > score1)
draw = 1 - p1_win - p2_win

print(f"Вероятность победы первого: {p1_win:.3f}")
print(f"Вероятность победы второго: {p2_win:.3f}")
print(f"Вероятность ничьей: {draw:.3f}")

# Графики распределений результатов за один раунд
plt.figure(figsize=(8,5))
plt.hist(score1, bins=50, alpha=0.6, label='Лучник 1 (Rayleigh(0,3))', density=True)
plt.hist(score2, bins=50, alpha=0.6, label='Лучник 2 (Rayleigh(5,1))', density=True)
plt.xlabel("Очки за раунд")
plt.ylabel("Плотность")
plt.title("Распределение очков за раунд у двух лучников")
plt.legend()
plt.grid(True)
plt.show()
```

## Задача 1

Случайные величины имеют нормальное распределение с $N \left[ \left(\begin{matrix}-5 \\ 2 \end{matrix}\right), \left(\begin{matrix}0.5 & 0.2\\ 0.2 & 0.4 \end{matrix}\right)\right]$

Необходимо найти математическое ожидание и дисперсию случайной величины $z=4X-5Y+25$.

*Подсказка*: стоит посмотреть на свойства мат. ожидания и дисперсии в «Википедии».

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Данные
mu = np.array([-2.5, 2.0])
Sigma = np.array([[0.5, 0.2],
                  [0.2, 0.4]])
a = np.array([4.0, -5.0])
c = 25.0

# Аналитика
E_z = a.dot(mu) + c
Var_z = a.dot(Sigma.dot(a))
std_z = np.sqrt(Var_z)

print("Аналитически:")
print(f"E[Z] = {E_z}")
print(f"Var(Z) = {Var_z}")

# Симуляция для проверки
Nsim = 200000
samples = np.random.multivariate_normal(mu, Sigma, size=Nsim)
z_samples = samples.dot(a) + c

print("\nВыборочные характеристики (симуляция):")
print(f"Среднее (выборочное) = {z_samples.mean():.4f}")
print(f"Дисперсия (выборочная) = {z_samples.var(ddof=0):.4f}")

# Построение гистограммы и аналитической плотности
t = np.linspace(E_z - 4*std_z, E_z + 4*std_z, 400)
pdf = norm.pdf(t, loc=E_z, scale=std_z)

plt.figure(figsize=(10,5))
plt.hist(z_samples, bins=180, density=True, alpha=0.35, label='Гистограмма выборки Z')
plt.plot(t, pdf, 'r-', lw=2, label=f'Аналитическая плотность N({E_z:.2f}, {Var_z:.2f})')
plt.xlabel('z')
plt.ylabel('Плотность')
plt.title('Распределение Z = 4X - 5Y + 25')
plt.legend()
plt.grid(True)
plt.show()
```

## Задача 2

В урне два белых и семь чёрных шаров. Из неё последовательно вынимают два шара. Какова вероятность того, что второй шар окажется белым **при условии,** что первый шар был чёрным?

```python
import random
import matplotlib.pyplot as plt

# Параметры
n_white = 2
n_black = 7
trials = 100_000

success = 0
results = []

for _ in range(trials):
    urn = ['W'] * n_white + ['B'] * n_black
    random.shuffle(urn)
    first = urn.pop(0)   # первый шар
    second = urn.pop(0)  # второй шар

    if first == 'B':  # условие: первый чёрный
        if second == 'W':
            success += 1
        results.append(second)

# Вероятность
probability = success / results.count('W') if results.count('W') != 0 else 0
theoretical = 2 / 8

print(f"Теоретическая вероятность: {theoretical:.4f}")
print(f"Эмпирическая вероятность: {success / len(results):.4f}")

# Визуализация
labels = ['Белый', 'Чёрный']
counts = [results.count('W'), results.count('B')]

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['white', 'black'], textprops={'color':"black"})
plt.title("Распределение второго шара при условии, что первый был чёрным")
plt.show()
```

## Задача 3

Найдите вероятность отказа схемы, предполагая, что отказы отдельных элементов независимы, а вероятность отказа элемента с номером равна $0.2$.

image.png

```python
import matplotlib.pyplot as plt

# Дано
p = 0.2  # отказ
q = 1 - p  # исправность

# Верхняя ветка: 1-2-3 (последовательно)
P_top = q**3

# Нижняя ветка: 4 - (5 || 6)
P_56 = 1 - (p * p)
P_bottom = q * P_56

# Параллель верх и низ
P_parallel = 1 - (1 - P_top) * (1 - P_bottom)

# Общая схема (после параллели идет 7)
P_system = P_parallel * q
P_fail = 1 - P_system

print(f"Вероятность исправности системы: {P_system:.4f}")
print(f"Вероятность отказа системы: {P_fail:.4f}")

# Визуализация
labels = ['Исправна', 'Отказ']
sizes = [P_system, P_fail]
colors = ['#66b3ff', '#ff6666']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Вероятность работы/отказа схемы')
plt.show()
```

2. Найдите вероятность того, что взятая стандартная деталь изготовлена вторым цехом.
*Подсказка*: теорема Байеса.

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
p_F1, p_F2 = 0.4, 0.6           # доли поставок
p_S_F1, p_S_F2 = 0.90, 0.95     # доля стандартных в цехах

# Теоретика
P_S = p_S_F1*p_F1 + p_S_F2*p_F2
P_F2_given_S = (p_S_F2*p_F2) / P_S

print(f"Теоретически: P(S) = {P_S:.4f}")
print(f"Теоретически: P(F2 | S) = {P_F2_given_S:.4f}")

# --- Монте-Карло для проверки ---
N = 1_000_000
src = np.random.choice([1,2], size=N, p=[p_F1, p_F2])
# стандартность по условной вероятности
u = np.random.rand(N)
is_standard = np.where(src==1, u < p_S_F1, u < p_S_F2)

P_S_emp = is_standard.mean()
P_F2_given_S_emp = ((src==2) & is_standard).sum() / is_standard.sum()

print(f"Эмпирически: P(S) ≈ {P_S_emp:.4f}")
print(f"Эмпирически: P(F2 | S) ≈ {P_F2_given_S_emp:.4f}")

# --- Визуализация ---
plt.figure(figsize=(11,4))

# 1) Сравнение
plt.subplot(1,2,1)
x = np.arange(2)
theory = [P_S, P_F2_given_S]
emp = [P_S_emp, P_F2_given_S_emp]
w = 0.35
plt.bar(x-w/2, theory, width=w, label='Теория')
plt.bar(x+w/2, emp, width=w, label='Моделирование')
plt.xticks(x, ['P(S)', 'P(F2 | S)'])
plt.ylim(0,1)
plt.ylabel('Вероятность')
plt.title('Сравнение теории и моделирования')
plt.grid(axis='y', alpha=0.3)
plt.legend()

# 2) Пирог только среди стандартных: из какого цеха
plt.subplot(1,2,2)
cnt_F1_S = ((src==1) & is_standard).sum()
cnt_F2_S = ((src==2) & is_standard).sum()
plt.pie([cnt_F1_S, cnt_F2_S],
        labels=['Стандартные от цеха 1','Стандартные от цеха 2'],
        autopct='%1.1f%%',
        colors=['#8ecae6','#ffb703'],
        startangle=90)
plt.title('Распределение стандартных деталей по цехам')

plt.tight_layout()
plt.show()
```
